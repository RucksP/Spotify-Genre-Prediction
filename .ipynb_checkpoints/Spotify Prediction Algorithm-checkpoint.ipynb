```python
# import any libraries we will use for this classification problem
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
from scipy import stats
from sklearn.cluster import KMeans
```

```python
# load the raw dataset
raw_data = pd.read_csv('Data/SpotifyFeatures.csv')
raw_data
```

```python
# Graph the raw data if we can
```

```python
# Clean the data of any weird or unneccessary categories
cleaned_data = pd.read_csv('Data/CleanSpotifyData.csv')
```

```python
# Normalize the Data
columns = list(cleaned_data.columns)
#remove the genre label from the data we will run k-means on
columns.remove('genre')
normalized_data = stats.zscore(cleaned_data[columns])
```

```python
# Run the K-Means Algorithm on the normalized Data

kmeans = KMeans(n_clusters=20, random_state=0).fit(normalized_data)
labels = kmeans.labels_

cleaned_data['clusters'] = labels
columns.extend(['clusters'])
```

```python
# the Dataframe of what songs are grouped to which clusters
clean_mean = cleaned_data.groupby(['clusters']).mean()
```
